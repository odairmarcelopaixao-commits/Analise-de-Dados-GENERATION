```python
import requests
import pandas as pd

# Indaiatuba
lat_br = -23.090
lon_br = -47.218

# Nova York
lat_us = 40.7128
lon_us = -74.0060

url = "https://api.open-meteo.com/v1/forecast"

params_br = {
    "latitude": lat_br,
    "longitude": lon_br,
    "daily": "temperature_2m_max,precipitation_sum",
    "timezone": "America/Sao_Paulo",
    "past_days": 7
}

params_us = {
    "latitude": lat_us,
    "longitude": lon_us,
    "daily": "temperature_2m_max,precipitation_sum",
    "timezone": "America/New_York",
    "past_days": 7
}

response_br = requests.get(url, params=params_br)
response_us = requests.get(url, params=params_us)

print(response_br.status_code)
print(response_us.status_code)

# JSON
data_br = response_br.json()
data_us = response_us.json()

# DataFrames
df_br = pd.DataFrame(data_br["daily"])
df_us = pd.DataFrame(data_us["daily"])

df_br.rename(columns={
    "time": "data",
    "temperature_2m_max": "temp_max_c",
    "precipitation_sum": "chuva_mm"
}, inplace=True)

df_us.rename(columns={
    "time": "data",
    "temperature_2m_max": "temp_max_c",
    "precipitation_sum": "chuva_mm"
}, inplace=True)

# Cidade
df_br["cidade"] = "Indaiatuba"
df_us["cidade"] = "New York"

# Fahrenheit
df_br["temp_f"] = (df_br["temp_max_c"] * 1.8) + 32
df_us["temp_f"] = (df_us["temp_max_c"] * 1.8) + 32

df_final = pd.concat([df_br, df_us], ignore_index=True)

df_final

# Salvar cvs como relatório.prof.luiz.csv
df_final.to_csv("relatorio_prof_luiz.csv", index=False)

from google.colab import files
files.download("relatorio_prof_luiz.csv")

df_final
```
